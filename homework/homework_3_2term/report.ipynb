### Опишем результаты графиков post enrichment 

- культура 3821:

*Воздействие интерфероном А* запускает мощную реакцию, которая меняет транскрипционный ландшафт клеток. Анализ показал активацию противовирусной защиты, что проявляется в индукции генов с РНК-связывающей и каталитической активностью. При этом ответ выходит за рамки чисто противовирусного: активация NOD-like рецепторного пути указывает на переход к активному провоспалительному ответу. Несмотря на это наблюдается сбалансированный ответ в виде механизмов негативной регуляции для предотвращения чрезмерного воспаления. Наиболее значимо то, что под действием интерферона клетки глиобластомы транскрипционно перепрограммируются, приобретая черты, характерные для тканей иммунной системы (лимфоидная ткань), что говорит об изменении их состояния в сторону более уязвимого для атак иммунной системы.  

При *воздействии интерфероном B* основной и наиболее значимый сигнал указывает на масштабную активацию генов, связанных со структурными компонентами клетки - промежуточными филаментами. В частности, основной эффект на данную культуру - реорганизация митохондрий, что указывает на запуск глубокой перестройки клетки, вероятно, связанной с дифференцировкой или альтернативным путем клеточного стресса. При этом отствует какое-либо упоминание противовируссного/иммунного ответа.   

- культура 4114:

Под *воздеййствием интерферона А* наблюадем несколько отличительных черт. Во-первых, наблюдается гиперактивация общего интерферонового сигналинга, в том числе специфических противовирусных систем. Во-вторых, противовирусный статус неразрывно связан с глубоким метаболическим перепрограммированием, проявляющимся в резкой активации окислительного фосфорилирования и комплексов дыхательной цепи (видим усиление энергопроизводства в митохондриях для поддержания иммунного ответа). В-третьих, показана активация сигнального пути IL-27  
  
*Воздействие интерфероном B* аналогично воздействию интерферона A: мощный проивовирусный ответ, метаболическое переключение на окислительное фосфорилирование и активация сигнального пути IL-27. 


- культура 6138:

Под *воздеййствием интерферона А* запускается классический противовирусный каскад, однако его особенностью является выраженная негативная регуляция специфических хемокинов (может целенаправленно подавлять хемокины, критически важные для привлечения иммунных клеток в микроокружение опухоли). При этом наблюдается и активация пути IL-27 и NOD-like рецепторов, что добавляет ответу сложности, сочетая иммуномодуляцию с провоспалительным потенциалом.  

*Воздействие интерфероном B* аналогично воздействию интерферона A, но отличие в отстутсвии негативной регуляции хемокинов IP-10 и CXCL2. При этом его воздействие вызывает более выраженную активацию митохондриальных компонентов и пути убиквитин-опосредованного протеолиза.

Метрики качества показывают, что интерфероновые отклики в трёх культурах глиобластомы различаются по силе и характеру:

- в культуре 4114 интерферон A вызывает наиболее мощный и достоверный ответ, тогда как интерферон B действует слабее и вовлекает намного меньше белков.

**Вывод:**     
1. Культура 3821 демонстрирует принципиально разные механизмы ответа на IFN-A и IFN-B. Под воздействием IFN-A наблюдается классический мощный иммунный ответ с перепрограммированием клеток в противовирусное состояние, активацией NOD-like рецепторов и приобретением черт иммунных тканей. В отличие от этого, IFN-B вызывает структурную перестройку клетки с активацией промежуточных филаментов и реорганизацией митохондрий без выраженного иммунного компонента. Метрики подтверждают сильный, но качественно различный ответ на оба интерферона.

2. Культура 4114 представляет собой гиперответчика со сходной реакцией на оба типа интерферона. Для этой линии характерна гиперактивация противовирусных систем с метаболическим переключением на окислительное фосфорилирование и устойчивой активацией пути IL-27. Однако метрики указывают, что IFN-A вызывает значительно более мощный и достоверный ответ по сравнению с IFN-B.

3. Культура 6138 показывает ответ с ключевым различием в регуляции хемокинов. Оба интерферона активируют противовирусную защиту и путь IL-27, но только IFN-A индуцирует негативную регуляцию хемокинов IP-10 и CXCL2, что может ограничивать привлечение иммунных клеток. IFN-B, в свою очередь, вызывает более выраженную митохондриальную перестройку. Метрики подтверждают слабый, но более согласованный ответ на IFN-B.

Таким образом, разные культуры глиобластомы показывают различные ответы на воздействие интерферонов, что отражает их индивидуальные биологические особенности.
